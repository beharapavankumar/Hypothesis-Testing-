# **Central Limit theorem :**

### **When a sufficiently large number of independent random samples are taken from a population, the sampling distribution of the sample mean will tend to be approximately normal, regardless of the population's original distribution, as long as the population has a finite variance.**

### **Key Points:**

**1)Applicability** : The theorem applies to many types of data distributions (e.g., skewed, uniform). 
It requires that:

The samples are independent.

The sample size is large (a common rule of thumb is n‚â•30).


**2)Mean and Variance:**

The mean of the sampling distribution of the sample mean is equal to the population mean (ùúá).

The variance of the sampling distribution is the population variance divided by the sample size (ùúé2/ùëõ).

# Example:

Suppose we are measuring the weight of apples in a large orchard. The weights are not normally distributed, and the distribution is right-skewed (many apples weigh less, but a few are much heavier).

**Steps to illustrate the CLT:**

**1.Population:** Assume the population has a mean weight of Œº=150 grams and a standard deviation of œÉ=30 grams.

**2.Sampling:** Draw random samples of size n=40. Compute the mean weight of each sample.

**3.Repeat Sampling:** Repeat the sampling process many times (e.g., 1000 times) to create a distribution of sample means.

**4.Result:**

Despite the original skewness of the weight distribution, the sampling distribution of the sample mean will appear bell-shaped (approximately normal).

The mean of this distribution will still be Œº=150, but the standard deviation (standard error) will shrink to ùúé/root(ùëõ)=30/root(40)‚âà4.74


**Applications:**

1.Hypothesis testing (e.g., z-tests, t-tests).

2.Confidence interval estimation.

3.Predicting probabilities using the normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters of the population distribution
population_mean = 150  # grams
population_std = 30    # grams
sample_size = 40
num_samples = 1000

# Generating a right-skewed population distribution (e.g., exponential shifted to fit the given mean and std)
population = np.random.exponential(scale=population_std / np.sqrt(2), size=100000) + (population_mean - population_std / np.sqrt(2))

# Sampling
sample_means = [np.mean(np.random.choice(population, sample_size, replace=True)) for _ in range(num_samples)]

# Visualization
plt.figure(figsize=(15, 8))

# Plot original population distribution
plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, kde=True, color='red', label='Population Distribution')
plt.axvline(population_mean, color='black', linestyle='--', label=f'Mean = {population_mean:.2f}')
plt.title('Population Distribution (Skewed)', fontsize=20)
plt.xlabel('Population mean Weight (grams)')
plt.ylabel('Frequency')
plt.legend()

# Plot sample mean distribution
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=50, kde=True, color='purple', label='Sample Mean Distribution')
plt.axvline(np.mean(sample_means), color='black', linestyle='--', label=f'Mean = {np.mean(sample_means):.2f}')
plt.title('Sampling Distribution of Sample Means', fontsize=20)
plt.xlabel('Sample Mean Weight (grams)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

The plots illustrate the Central Limit Theorem:

**Left Plot**: The original population distribution is skewed (right-skewed in this case), reflecting the inherent variability in the weights of apples.

**Right Plot**: The sampling distribution of the sample means (computed from multiple samples of size n=40) is approximately normal. Despite the skewness in the population, the sampling distribution is bell-shaped, centered around the population mean (Œº=150).

This demonstrates how the CLT works to normalize the distribution of sample means, enabling us to apply statistical methods that assume normality. 

# **Confidence Interval :**

A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the population parameter (such as the mean or proportion) with a certain level of confidence. It provides a way to estimate the uncertainty associated with a sample statistic.

### **Key Concepts:**

**1.Confidence Level:**

It represents the probability that the confidence interval contains the true population parameter if we repeated the sampling process many times.

Common confidence levels are 90%, 95%, and 99%.

**2.Structure of Confidence Interval:**

For the population mean (Œº), the confidence interval is typically calculated as:
CI= x ¬± ùëß ‚ãÖ ùúé/root(ùëõ)

Where:

ùë•: Sample mean.

ùëß: Critical value from the standard normal distribution (e.g., z=1.96 for 95% confidence).

ùúé: Population standard deviation (if unknown, we use the sample standard deviations (s)).

n: Sample size.

3.**Interpretation:**

A 95% confidence interval means that if we were to take 100 random samples and calculate the CI for each, about 95 of those intervals would contain the true population parameter.

# Example

Suppose we are studying the average height of adult males in a city. A random sample of 50 individuals yields:

Sample mean (ùë•): 175 cm.

Sample standard deviation (s): 10 cm.

We want to construct a 95% confidence interval for the population mean height.

```python
# Given data
sample_mean = 175  # cm
sample_std = 10     # cm
sample_size = 50    # n
z_value = 1.96      # Critical value for 95% confidence

# Calculate margin of error
margin_of_error = z_value * (sample_std / np.sqrt(sample_size))

# Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

lower_bound, upper_bound
```

The 95% confidence interval for the population mean height is approximately: **(172.23cm,177.77cm)**

**Interpretation:**

We are 95% confident that the true average height of adult males in the city lies between 172.23 cm and 177.77 cm.

This interval reflects the uncertainty due to sampling variability. If we repeated this study many times, 95% of the intervals would include the true mean.
